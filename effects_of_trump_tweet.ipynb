```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tweets.csv')
df
```

```python
df.dtypes
```

```python
df['시간'] = pd.to_datetime(df['created_at'], errors='coerce')
```

```python
df
```

```python
df.dtypes
```

```python
df['year'] = df['시간'].dt.year
df['month'] = df['시간'].dt.month
df['day'] = df['시간'].dt.day
df['date'] = df['시간'].dt.date

df['hour'] = df['시간'].dt.hour
df['minute'] = df['시간'].dt.minute
df['second'] = df['시간'].dt.second
df['time'] = df['시간'].dt.time
```

```python
df
```

```python
df.isnull().count()
```

```python
df_1 = df.drop(['created_at', 'id_str'], axis = 1)
df_1
```

```python
frequency = df_1['date'].value_counts()
```

```python
grouped_df = df_1.groupby('date').count()
grouped_df
```

```python
frequency_text = grouped_df[['text']]
frequency_text.head()
```
