```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import re as re
from textblob import TextBlob
import plotly.express as px
```

```python
df = pd.read_csv('tweets.csv')
df
```

```python
df_1 = df.drop(['id_str'], axis=1)
df_1
```

```python
df_1['시간'] = pd.to_datetime(df['created_at'], errors='coerce')
```

```python
df_1
```

```python
df_1.dtypes
```

```python
df_1['date'] = df_1['시간'].dt.date
```

```python
df_1
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub('(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)', ' ', tweet).split())

df_1["Clean Tweet"] = df_1['text'].apply(lambda x: clean_tweet(x))
```

```python
df_1
```

```python
df_1.to_pickle("cleaning_df.pkl")
```

```python
def analyze_sentiment(tweet):
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity == 0:
        return "Neutral"
    else:
        return "Negative"
```

```python
df_1["Sentiment"] = df_1["Clean Tweet"].apply(lambda x: analyze_sentiment(x))
df_1
```

```python
df_sen=df_1['Sentiment'].value_counts()
df_sen
```

```python
fig = px.bar(df_1, x="Sentiment", 
             color='Sentiment', labels={'Count':'Count'}, height=400, title="트럼프 트위터 감성분석")
fig.show()
```

```python
df_2 = df_1.groupby(by=['date'], as_index=False).count()
df_2
```

```python
fig = px.line(df_2, x="date", y="text", title="트럼프 하루 트윗 수(리트윗 포함)")
fig.show()
```
